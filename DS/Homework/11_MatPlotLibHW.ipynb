## Exercise 1: Read Total profit of all months and show it using a line plot

## Exercise 3: Read all product sales data and show it using a multiline plot
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

```python
data = pd.read_csv('company_sales_data.csv')
fig = plt.figure()
plt.plot(data.month_number, data.facecream, label='Facecream')
plt.plot(data.month_number, data.facewash, label='Facewash')
plt.plot(data.month_number, data.toothpaste, label='Toothpaste')
plt.plot(data.month_number, data.bathingsoap, label='Bath Soap')
plt.plot(data.month_number, data.shampoo, label='Shampoo')
plt.plot(data.month_number, data.moisturizer, label='Moisturizer')
plt.plot(data.month_number, data.total_units, label='Total Units')
plt.legend()
title = plt.title('Product Sales')
```

```python
#Erin McConnell's Group ALternative Solution
x = data['month_number']

colors = ['green', 'red', 'blue', 'yellow', 'purple', 'black']
markers = ['o', '.', ',', 'x', '+', 'v']
labels = ['Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer']

for i in range(1, 7):
    plt.plot(x, data[data.columns[i]], marker=markers[i-1], color=colors[i-1], label=labels[i-1])
    
plt.xlabel("Month Number")
plt.ylabel("Number of Units")
plt.legend()
```

## Exercise 5: Read face cream and face wash product sales data and show it using the bar chart
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```python
data = pd.read_csv('company_sales_data.csv')
plt.style.use('classic')
plt.bar(data.month_number, data.facecream, label='Face Cream', width = .375)
plt.bar(data.month_number+.375, data.facewash, label='Face Wash', width = .375)
plt.legend()
title = plt.title('Monthly Units Sold')
```
